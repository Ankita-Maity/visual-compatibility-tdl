```python
import re
import matplotlib.pyplot as plt
```

```python
with open('values/350_1.txt') as f:
    normal = f.readlines()
```

```python
with open('values/deg2.txt') as f:
    deg2 = f.readlines()
```

```python
with open('values/deg3.txt') as f:
    deg3 = f.readlines()
```

```python
epochs = []
train_acc = []
val_acc = []
for i in range (len(normal)):
    nums = re.findall(r'[-+]?[.]?[\d]+(?:,\d\d\d)*[\.]?\d*(?:[eE][-+]?\d+)?', normal[i])
    epochs.append(int(nums[0]))
    train_acc.append(float(nums[2]))
    val_acc.append(float(nums[4]))
```

```python
epochs2 = []
train_acc2 = []
val_acc2 = []
for i in range (len(deg2)):
    nums = re.findall(r'[-+]?[.]?[\d]+(?:,\d\d\d)*[\.]?\d*(?:[eE][-+]?\d+)?', deg2[i])
    epochs2.append(int(nums[0]))
    train_acc2.append(float(nums[2]))
    val_acc2.append(float(nums[4]))
```

```python
epochs3 = []
train_acc3 = []
val_acc3 = []
for i in range (len(deg3)):
    nums = re.findall(r'[-+]?[.]?[\d]+(?:,\d\d\d)*[\.]?\d*(?:[eE][-+]?\d+)?', deg3[i])
    epochs3.append(int(nums[0]))
    train_acc3.append(float(nums[2]))
    val_acc3.append(float(nums[4]))
```

```python
plt.figure(figsize=(9,6))
plt.plot(epochs, train_acc, label = "Train accuracy (degree 1)", color='wheat')
plt.plot(epochs2, train_acc2, label = "Train accuracy (degree 2)", color='orange')
plt.plot(epochs3, train_acc3, label = "Train accuracy (degree 3)", color='orangered')
plt.plot(epochs, val_acc, label = "Validation accuracy (degree 1)", color='aqua')
plt.plot(epochs2, val_acc2, label = "Validation accuracy (degree 2)", color='mediumblue')
plt.plot(epochs3, val_acc3, label = "Validation accuracy (degree 3)", color='midnightblue')
plt.legend()
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.show()
```
