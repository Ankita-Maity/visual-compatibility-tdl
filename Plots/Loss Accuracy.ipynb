```python
import re
import matplotlib.pyplot as plt
```

```python
with open('values/values.txt') as f:
    lines = f.readlines()
```

```python
epochs = []
train_loss = []
val_loss = []
train_acc = []
val_acc = []
for i in range (len(lines)):
    nums = re.findall(r'[-+]?[.]?[\d]+(?:,\d\d\d)*[\.]?\d*(?:[eE][-+]?\d+)?', lines[i])
    epochs.append(int(nums[0]))
    train_loss.append(float(nums[1]))
    train_acc.append(float(nums[2]))
    val_loss.append(float(nums[3]))
    val_acc.append(float(nums[4]))
```

```python
plt.figure(figsize=(9,6))
plt.plot(epochs, train_loss, label = "Train loss")
plt.plot(epochs, val_loss, label = "Validation loss")
plt.legend()
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.show()
```

```python
plt.figure(figsize=(9,6))
plt.plot(epochs, train_acc, label = "Train accuracy")
plt.plot(epochs, val_acc, label = "Validation accuracy")
plt.legend()
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.show()
```
